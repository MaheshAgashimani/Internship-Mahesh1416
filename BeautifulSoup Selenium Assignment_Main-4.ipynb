```python
Series_name=[]
Series_Year_Span=[]
Series_Genre=[]
Series_Run_Time=[]
Series_Rating=[]
Series_Vote=[]


Series_name_tags=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']")
for i in Series_name_tags:
    #time.sleep(5)
    try:
        Series_name.append(i.text)
    except NoSuchElementException :
        Series_name.append('Not present')

Series_Year_Span_tags=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
for i in Series_Year_Span_tags:
    #time.sleep(5)
    try:
        Series_Year_Span.append(i.text)
    except NoSuchElementException :
        Series_Year_Span.append('Not present')
        
Series_Genre_tags=driver.find_elements(By.XPATH,"//span[@class='genre']")
for i in Series_Genre_tags:
    #time.sleep(5)
    try:
        Series_Genre.append(i.text)
    except NoSuchElementException :
        Series_Genre.append('Not present')
        
Series_Run_Time_tags=driver.find_elements(By.XPATH,"//span[@class='runtime']")
for i in Series_Run_Time_tags:
    #time.sleep(5)
    try:
        Series_Run_Time.append(i.text)
    except NoSuchElementException :
        Series_Run_Time.append('Not present')
        
Series_Rating_tags=driver.find_elements(By.XPATH,"//div[@class='ipl-rating-star small']")
for i in Series_Rating_tags:
    #time.sleep(5)
    try:
        Series_Rating.append(i.text)
    except NoSuchElementException :
        Series_Rating.append('Not present')
        
Series_Vote_tags=driver.find_elements(By.XPATH,"//p[@class='text-muted text-small']")
for i in Series_Vote_tags:
    #time.sleep(5)
    try:
        Series_Vote.append(i.text[2].split(","))
    except NoSuchElementException :
        Series_Vote.append('Not present')
```

```python
print(len(Series_name),len(Series_Year_Span),len(Series_Genre),len(Series_Run_Time),len(Series_Rating),len(Series_Vote))
```

```python
df=pd.DataFrame({'Series_name':Series_name,'Series_Year_Span':Series_Year_Span,'Series_Genre':Series_Genre,'Series_Run_Time':Series_Run_Time,'Series_Rating':Series_Rating})
df
```

```python
pop_up=driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
pop_up.click()
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Instances=[]
Attributes=[]
Year=[]


Dataset_Name_tags=driver.find_elements(By.XPATH,"//p[@class='normal']//b//a")
for i in Dataset_Name_tags:
    #time.sleep(5)
    try:
        Dataset_Name.append(i.text)
    except NoSuchElementException :
        Dataset_Name.append('Not present')

Data_Type_tags=driver.find_elements(By.XPATH,"//table[@class='cellpadding']//tbody//tr//td[2]//table[2]//tbody//tr[2]//td[1]//table//tbody//tr//td[3]")
for i in Data_Type_tags:
    #time.sleep(5)
    try:
        Data_Type.append(i.text)
    except NoSuchElementException :
        Data_Type.append('Not present')
        
Task_tags=driver.find_elements(By.XPATH,"//table[@class='cellpadding']//tbody//tr//td[2]//table[2]//tbody//tr[2]//td[1]//table//tbody//tr//td[3]")
for i in Task_tags:
    #time.sleep(5)
    try:
        Task.append(i.text)
    except NoSuchElementException :
        Task.append('Not present')
        
Instances_tags=driver.find_elements(By.XPATH,"//table[@class='cellpadding']//tbody//tr//td[2]//table[2]//tbody//tr[2]//td[1]//table//tbody//tr//td[4]")
for i in Instances_tags:
    #time.sleep(5)
    try:
        Instances.append(i.text)
    except NoSuchElementException :
        Instances.append('Not present')
        
Attributes_tags=driver.find_elements(By.XPATH,"//table[@class='cellpadding']//tbody//tr//td[2]//table[2]//tbody//tr[2]//td[1]//table//tbody//tr//td[5]")
for i in Attributes_tags:
    #time.sleep(5)
    try:
        Attributes.append(i.text)
    except NoSuchElementException :
        Attributes.append('Not present')
        
Year_tags=driver.find_elements(By.XPATH,"//table[@class='cellpadding']//tbody//tr//td[2]//table[2]//tbody//tr[2]//td[1]//table//tbody//tr//td[6]")
for i in Year_tags:
    #time.sleep(5)
    try:
        Year.append(i.text)
    except NoSuchElementException :
        Year.append('Not present')
```

```python
print(len(Dataset_Name),len(Data_Type),len(Task),len(Instances),len(Attributes),len(Year))
```

```python
df=pd.DataFrame({'Dataset_Name':Dataset_Name,'Data_Type':Data_Type,'Task':Task,'Instances':Instances,'Attributes':Attributes,'Year':Year})
df
```
